```python
dirs = local.data_directory()
qobj, Qobj = {}, {}
for wind, ivar, ascale in [('com16','noivar',3.),('iso15','V3',3.),('iso15','base',3.),('com15','noivar',1.)]:
    aobj = local.init_analysis_params(qid='comb_d',ascale=ascale,wind=wind,ivar=ivar)
    qobj[wind+ivar] = quad_func.reconstruction(dirs['local'],aobj.ids,stag=aobj.stag,rlmin=500,qlist=['TT'],bhe=['src'])
    Qobj[wind+ivar] = quad_func.reconstruction(dirs['local'],aobj.ids,stag=aobj.stag,rlmin=500,qlist=['TT'])
```
